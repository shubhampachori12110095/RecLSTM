```python
import tensorflow as tf
from datapreprocessing import lyx_batch_generator
import rec_lstm
import os
import pickle


def start_train():
    num_to_item_pkl = 'data/num_to_item.pkl'
    with open(num_to_item_pkl, 'rb') as rf:
        num_to_item_dict = pickle.load(rf)
    train_flags = {'name': 'lstm_lyx',
                   'num_items': (len(num_to_item_dict)+1),
                   'num_seqs': 64,
                   'num_steps': 30,
                   'lstm_size': 32,
                   'num_layers': 3,
                   'learning_rate': 0.001,
                   'train_keep_prob': 0.5,
                   'max_steps': 30000,
                   'save_every_n': 1000,
                   'log_every_n': 10
                  }
    model_path = os.path.join('model', train_flags['name'])
    if os.path.exists(model_path) is False:
        os.makedirs(model_path)
    g = lyx_batch_generator(train_flags['num_seqs'], train_flags['num_steps'])
    model = rec_lstm.RecLSTM(train_flags['num_items'],
                             num_seqs=train_flags['num_seqs'],
                             num_steps=train_flags['num_steps'],
                             lstm_size=train_flags['lstm_size'],
                             num_layers=train_flags['num_layers'],
                             learning_rate=train_flags['learning_rate'],
                             train_keep_prob=train_flags['train_keep_prob']
                            )
    model.train(g,
                train_flags['max_steps'],
                model_path,
                train_flags['save_every_n'],
                train_flags['log_every_n'])
            
            
if __name__ == '__main__':
    start_train()
```
