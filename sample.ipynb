```python
import tensorflow as tf
import rec_lstm
import os
import pickle
from numpy import random as nr
import numpy as np


input_target_list_pkl = 'data/input_taget_list.pkl'
num_to_item_pkl = 'data/num_to_item.pkl'


def lyx_sample_generator(n_steps):
    with open(input_target_list_pkl, 'rb') as rf:
        input_target_list = pickle.load(rf)
#     print(len(input_target_list))
    index = 0
    while True:
        if index < len(input_target_list):
            x = list()
#             random_index = nr.randint(0, len(input_target_list))
            sample_list = input_target_list[index]
            data_length = len(sample_list) - 1
            zero_length = n_steps - data_length
            zero_list = []
            for _ in xrange(zero_length):
                zero_list.append(0)
            x.append(sample_list[:-1] + zero_list)
            index += 1
            yield x
        else:
            break


def start_sample():
    with open(num_to_item_pkl, 'rb') as rf:
        num_to_item_dict = pickle.load(rf)
    sample_flags = {'num_items': (len(num_to_item_dict)+1),
                    'num_seqs': 1,
                    'num_steps': 30,
                    'lstm_size': 32,
                    'num_layers': 3,
                    'checkpoint_path': 'model/lstm_lyx/'
                   }
    if os.path.isdir(sample_flags['checkpoint_path']):
        sample_flags['checkpoint_path'] = tf.train.latest_checkpoint(
            sample_flags['checkpoint_path'])
    model = rec_lstm.RecLSTM(sample_flags['num_items'],
                             num_seqs=sample_flags['num_seqs'],
                             num_steps=sample_flags['num_steps'],
                             lstm_size=sample_flags['lstm_size'],
                             num_layers=sample_flags['num_layers']
                            )
    model.load(sample_flags['checkpoint_path'])
    g = lyx_sample_generator(sample_flags['num_steps'])
    model.test(g, sample_flags['num_items'], 600)

        
if __name__ == '__main__':
    start_sample()
```
